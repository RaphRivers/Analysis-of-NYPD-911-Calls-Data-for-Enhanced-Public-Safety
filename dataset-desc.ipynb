## Calls for Service to NYPD's 911 system dataset description

This dataset compiles information from the NYPD 911 system, known as ICAD. It gathers data from the ICAD system, which call takers and dispatchers use to interact with both callers from the public and the NYPD itself. Each record in the dataset corresponds to an entry in the system, encompassing reports from the general public and self-initiated entries made by NYPD Members of Service. This data is valuable for understanding and addressing matters handled by the NYPD.

### Dataset Source Information:
[NYC Open Data](https://data.cityofnewyork.us/Public-Safety/NYPD-Calls-for-Service-Year-to-Date-/n2zq-pubd/about_data)
